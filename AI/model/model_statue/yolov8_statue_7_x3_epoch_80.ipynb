```python
# ÏÇ¨Ï†Ñ ÌïôÏäµ Î™®Îç∏ Îã§Ïö¥Î°úÎìú
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
```

```python
# Ï†ÑÏù¥ ÌïôÏäµ
model.train(data='/root/.jupyter/yolo/statue_7/statue_7_yaml.yaml', epochs=80, patience=80)
```

```python
print(type(model.names), len(model.names))
print(model.names)
```

```python
results = model.predict(source='/root/.jupyter/yolo/statue_7/test/images', save=True)
```

```python
print(type(results), len(results))
```

```python
import numpy as np

detected_num = 0

for result in results:
    
    uniq, cnt = np.unique(result.boxes.cls.cpu().numpy(), return_counts=True)
    uniq_cnt_dict = dict(zip(uniq, cnt))
    
    print('\n{class num:counts} =', uniq_cnt_dict, '\n')
    
    if uniq is not None and uniq.size > 0 and len(set(uniq)) == 1:
        detected_num += 1
    
    for c in result.boxes.cls:
        print('class num=', int(c), ' , class_name =', model.names[int(c)])
```
