```python
import yaml
```

```python
# 커스텀 yaml 파일 생성
# yamle 파일은 트레이닝 데이터와 테스트 데이터의 메타 데이터 정보를 가지고 있다.

data = { 'train' : r'\Users\YoungHan\Desktop\yolov8\buildings_12\train\images',
         'val' : r'\Users\YoungHan\Desktop\yolov8\buildings_12\valid\images',
         'test' : r'\Users\YoungHan\Desktop\yolov8\buildings_12\test\images',
         'names' : ['b105', 'b107', 'b108', 'b201', 'b205', 'b206', 'b207', 'b208', 'b303', 'b306', 'b311', 'b312'],
         'nc' : 12 }

with open(r'\Users\YoungHan\Desktop\yolov8\buildings_12\buildings_12_yaml.yaml', 'w') as f:
    yaml.dump(data, f)

with open(r'\Users\YoungHan\Desktop\yolov8\buildings_12\buildings_12_yaml.yaml', 'r') as f:
    building_yaml = yaml.safe_load(f)
    display(building_yaml)
```

```python
import ultralytics
ultralytics.checks()
```

```python
# 사전 학습 모델 다운로드
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
```

```python
print(type(model.names), len(model.names))
print(model.names)
```

```python
# 전이 학습
model.train(data=r'\Users\YoungHan\Desktop\yolov8\buildings_12\buildings_12_yaml.yaml', epochs=10)
```

```python
print(type(model.names), len(model.names))
print(model.names)
```

```python
results = model.predict(source=r'\Users\YoungHan\Desktop\yolov8\buildings_12\test\images', save=True)
```

```python
print(type(results), len(results))
```

```python
import numpy as np

detected_num = 0

for result in results:
    
    uniq, cnt = np.unique(result.boxes.cls.cpu().numpy(), return_counts=True)
    uniq_cnt_dict = dict(zip(uniq, cnt))
    
    print('\n{class num:counts} =', uniq_cnt_dict, '\n')
    
    if uniq is not None and uniq.size > 0 and len(set(uniq)) == 1:
        detected_num += 1
    
    for c in result.boxes.cls:
        print('class num=', int(c), ' , class_name =', model.names[int(c)])
```
